```python
with open('receipts.txt', 'r') as file:
    data = file.read().replace('\n', '')
data
```

```python
pattern = r"Quantity +\d+"
re.findall(pattern, data)
result = []
for s in re.findall(pattern, data):
    result.append(re.findall(r"\d+",s))
result[0:3]
```

```python
pattern = r"Total Amount +\d+\.\d+"
re.findall(pattern, data)
result = []
for s in re.findall(pattern, data):
    result.append(re.findall(r"\d+\.\d+",s))
result[0:3]
```

```python
pattern = r"Quantity +(\d+)"
re.findall(pattern, data)
```

```python
pattern = r"Total Amount +(\d+\.\d+)"
re.findall(pattern, data)
```

```python
receipts = data.split(" â‚¬====================================************************************")
receipts
```

```python
receipts_dict = {
    "date": [],
    "quantity": [],
    "total_amount": []
}
for r in receipts:
    date_regexp = re.findall(r"\d{2}-\d{2}-\d{4}",r)[0]
    receipts_dict["date"].append(date_regexp)
    quantity_regexp = re.findall(r"Quantity +(\d+)",r)
    receipts_dict["quantity"].append(quantity_regexp[0])
    amount_regexp = re.findall(r"Total Amount +(\d+.\d{2})",r)
    receipts_dict["total_amount"].append(amount_regexp[0])
receipts_dict
```

```python
receipt_df = pd.DataFrame.from_dict(receipts_dict)
receipt_df["date"]=pd.to_datetime(receipt_df["date"])
receipt_df["quantity"]=pd.to_numeric(receipt_df["quantity"])
receipt_df["total_amount"]=pd.to_numeric(receipt_df["total_amount"])
receipt_df.sort_values(["date"],ascending=True, inplace=True) 
receipt_df.set_index("date", inplace=True)
receipt_df.plot(figsize=(15,5)) 
```
